```python
import tkinter as tk
from tkinter import simpledialog, messagebox

class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        messagebox.showinfo("Success", f"Book '{book.title}' added to the library.")

    def display_books(self):
        if not self.books:
            messagebox.showinfo("Empty Library", "No books in the library.")
        else:
            book_list = "\n".join([f"Title: {book.title}, Author: {book.author}, ISBN: {book.isbn}" for book in self.books])
            messagebox.showinfo("Books in Library", book_list)

    def search_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                messagebox.showinfo("Book Found", f"Book found - Title: {book.title}, Author: {book.author}, ISBN: {book.isbn}")
                return
        messagebox.showinfo("Book Not Found", f"Book with title '{title}' not found in the library.")

class LibraryGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Library Management System")

        self.library = Library()

        self.label = tk.Label(root, text="Library Management System", font=("Helvetica", 16))
        self.label.pack(pady=10)

        self.add_button = tk.Button(root, text="Add Book", command=self.add_book)
        self.add_button.pack()

        self.display_button = tk.Button(root, text="Display Books", command=self.display_books)
        self.display_button.pack()

        self.search_button = tk.Button(root, text="Search Book", command=self.search_book)
        self.search_button.pack()

        self.exit_button = tk.Button(root, text="Exit", command=root.destroy)
        self.exit_button.pack()

    def add_book(self):
        title = simpledialog.askstring("Input", "Enter book title:")
        author = simpledialog.askstring("Input", "Enter author name:")
        isbn = simpledialog.askstring("Input", "Enter ISBN:")
        new_book = Book(title, author, isbn)
        self.library.add_book(new_book)

    def display_books(self):
        self.library.display_books()

    def search_book(self):
        title = simpledialog.askstring("Input", "Enter book title to search:")
        self.library.search_book(title)

if __name__ == "__main__":
    root = tk.Tk()
    app = LibraryGUI(root)
    root.mainloop()
```
